### Ejercicio. Equipos

Se tienen dos listas que contienen los nombres de los jugadores de dos equipos diferentes.
Las listas están predefinidas con los siguientes valores:

```python
jugadores1=["JUAN","ANA","LUZ","TOMAS","RAUL","SARA","LUIS","BORGA","RAFA"]
jugadores2=["PAM","CRUZ","ARON","GONZALO","TATO","LUIS","JUAN","TORO","RAFA","ANITA","CARLA","JUAN" ]
```

El programa debe realizar los siguientes análisis e imprimir:
1. Los nombres de todos los jugadores (de ambos equipos) que contengan al menos una de las vocales: "A", "E" o "I".
2. Los nombres de todos los jugadores (de ambos equipos) que contengan dos o más veces la misma vocal.
3. Los nombres de los jugadores que aparezcan en ambos equipos (nombres repetidos).

````python
jugador1 = ["JUAN","ANA","LUZ","TOMAS","RAUL","SARA","LUIS","BORGA","RAFA"]
jugador2 = ["PAM","CRUZ","ARON","GONZALO","TATO","LUIS","JUAN","TORO","RAFA","ANITA","CARLA","JUAN"]
jugadores_todos = jugador1 + jugador2
lista_AEI = []
lista_vocales = []
lista_repetidos = []
vocales = ['A', 'E', 'I', 'O', 'U']

# 1. Jugadores con A, E, I
for jugador in jugadores_todos:
    for letra in jugador:
        if letra in 'AEI':
            if jugador not in lista_AEI: # evita repetidos
                lista_AEI.append(jugador)

# 2. Jugadores con vocales repetidas
for jugador in jugadores_todos:
    for vocal in vocales:
        # Contar cuántas veces aparece cada vocal
        cantidad = 0
        for letra in jugador:
            if letra == vocal:
                cantidad += 1
        # Si alguna vocal aparece 2 o más veces
        if cantidad >= 2:
            if jugador not in lista_vocales:
                lista_vocales.append(jugador)

# 3. Nombres repetidos entre equipos
for jugador in jugador1:
    for nombre in jugador2:
        if jugador == nombre:
            if jugador not in lista_repetidos:
                lista_repetidos.append(jugador)

print('Jugadores con A, E o I en su nombre:')
print(lista_AEI)
print('Jugadores con vocales repetidas:')
print(lista_vocales)
print('Jugadores con nombres repetidos:')
print(lista_repetidos)
````

````python
jugador1=["JUAN","ANA","LUZ","TOMAS","RAUL","SARA","LUIS","BORGA","RAFA"]
jugador2=["PAM","CRUZ","ARON","GONZALO","TATO","LUIS","JUAN","TORO","RAFA","ANITA","CARLA","JUAN" ]
jugadores=jugador1 + jugador2
lista_con_AEI=[]
lista_iguales=[]
lista_repetidos=[]
vocales=['A','E','I','O','U']

for jugador in jugadores:
    if 'A' in jugador or 'E' in jugador or 'I' in jugador: # si el jugador tiene una A o una E o una I.
        lista_con_AEI += [jugador]

for jugador in jugadores:
    for v in vocales:
        if jugador.count(v)>1:
            lista_iguales += [jugador]
            break # encuentro dos vocales, no busco si hay más, con dos vocales al menos, vale.
            
for jugador in jugadores:
    if jugadores.count(jugador)>1:
        if lista_repetidos.count(jugador)==0:
            lista_repetidos += [jugador]
        
print('Jugadores de ambos equipos')
print(jugadores)
print('Jugadores con las vocales  A E I en el nombre')
print(lista_con_AEI)
print('Jugadores con dos vocales o más iguales en el nombre  ')
print(lista_iguales)
print('Nombres de jugadores repetidos en ambos equipos')
print(lista_repetidos)
````
